```python
#Data PreProcessing
#splitting data into dependent and independent vaiables
X=data.iloc[:,:6]
y=data.iloc[:,6]
```

```python
print("Shape of x: ",X.shape)
print("Shape of y: ", y.shape)
```

```python
# splitting the dataset into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 0)
```

```python
# standardization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Part 3: Modeling

```python
# import libraries for metrics and reporting
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import log_loss
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, accuracy_score
```

```python
n_splits = 5
kf  = KFold(n_splits =n_splits, shuffle=True, random_state=100)
```

```python
# Hyperparameters to be checked.
parameters_lg = {'C':[0.0001,0.001, 0.01, 1, 0.1, 10, 100, 1000],
              'penalty':['none','l2'],
              'solver':['lbfgs','sag','saga','newton-cg']
             }

# Logistic Regression classifier.
default_logreg=LogisticRegression(multi_class='auto', random_state=42, n_jobs=-1)

# GridSearchCV estimator.
gs_logreg = GridSearchCV(default_logreg, parameters_lg, cv=kf, verbose=1)
```

```python
# Train the GridSearchCV estimator and search for the best parameters.
gs_logreg.fit(X_train,y_train)
```

```python
# Make predictions with the best parameters.
y_pred=gs_logreg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
# Best parameters.
print("Best Logistic Regression Parameters: {}".format(gs_logreg.best_params_))

# Cross validation accuracy for the best parameters.
print('CV Accuracy: %0.3f' % gs_logreg.best_score_)

# Accuracy: 1 is perfect prediction.
print('Accuracy: %0.3f' % (gs_logreg.score(X_test,y_test)))

cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
# Hyperparameters to be checked.
#parameters_nb ={'priors': None, 'var_smoothing': [1e-07,1e-08,1e-09]}
parameters_nb ={'var_smoothing': [1e-07,1e-08,1e-09]}
# Default NB classifier.
default_nb=GaussianNB()

# GridSearchCV estimator.
gs_nb = GridSearchCV(default_nb, parameters_nb, cv=kf, verbose=1)
# Train the GridSearchCV estimator and search for the best parameters.
gs_nb.fit(X_train,y_train)
# Make predictions with the best parameters.
y_pred=gs_nb.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
# Best parameters.
print("Best NB Parameters: {}".format(gs_nb.best_params_))

# Cross validation accuracy for the best parameters.
print('CV Accuracy: %0.3f' % gs_nb.best_score_)

# Accuracy: 1 is perfect prediction.
print('Accuracy: %0.3f' % (gs_nb.score(X_test,y_test)))

cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
